```python
from pydantic import BaseModel,Field

class Movie(BaseModel):
    name: str = Field(description="The name of the movie")
    year: int = Field(description="The year the movie was released")
    genre: str = Field(description="The genre of the movie")
    rating: float = Field(description="The rating of the movie")
```

```python
import os
import dotenv
dotenv.load_dotenv()

from langchain_groq import ChatGroq
os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')
```

```python
llm = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=0,
    max_tokens=None,
    reasoning_format="parsed",
    timeout=None,
    max_retries=2,
)
```

```python
model_with_structure = llm.with_structured_output(Movie)

model_with_structure
```

```python
llm.invoke("what is best movie of nepal?")
```

```python
model_with_structure.invoke("explain movie inception?")
```

```python
model_with_structure.invoke("what is best movie of nepal?")

```

```python
class Movie(BaseModel):
    name: str = Field(...,description="The name of the movie")
    year: int = Field(...,description="The year the movie was released")
    genre: str = Field(...,description="The genre of the movie")
    rating: float = Field(...,description="The rating of the movie")
```

```python
model_with_structure = llm.with_structured_output(Movie,include_raw=True)
model_with_structure.invoke("what is best movie of nepal?")
```

```python
## Nested Structure

class Actor(BaseModel):
    name:str
    age:str

class MovieDetail(BaseModel):
    actors:list[Actor]
    title:str
    released:str
    genre: list[str]
    budget: float | Field(None,description="Budget in million USD")


```

```python
model_with_structure = llm.with_structured_output(MovieDetail,include_raw=True)
model_with_structure.invoke("explain movie inception")
```

```python
class contact (BaseModel):
    name:str = Field(description="name of user")
    email:str = Field(description="email of user")
```

```python
#lets work with agents
from langchain.agents import create_agent
agent = create_agent(
    llm,
    response_format = contact
)

agent.invoke({'messages':[{
    "role":"user",
    "content":"add john doe email johndoe23@gmail.com"
}]})
```
